```python
def calculate_trade(option_row, option_type):

    day = option_row['Date']
    ticker = option_row['Ticker']

    entry_price = option_row['Close']

    stoploss = entry_price * 1.5

    option_data = df[
        (df['Date'] == day) &
        (df['Ticker'] == ticker)
    ]

    after_entry = option_data[
        option_data['Time'] > '09:20:59'
    ]

    sl_hit = after_entry[
        after_entry['High'] >= stoploss
    ]

    if len(sl_hit) > 0:

        exit_time = sl_hit.iloc[0]['Time']
        exit_price = stoploss

    else:

        exit_row = option_data[
            option_data['Time'] == '15:20:59'
        ]

        if len(exit_row) == 0:
            return None

        exit_time = '15:20:59'
        exit_price = exit_row.iloc[0]['Close']

    pnl = (entry_price - exit_price) * 15

    return {
        'Date': day,
        'Ticker': ticker,
        'Type': option_type,
        'EntryPrice': entry_price,
        'ExitPrice': exit_price,
        'ExitTime': exit_time,
        'PnL': pnl
    }
```

```python
from tqdm import tqdm

ticker_lookup = {}

for key, group in tqdm(df.groupby(['Date', 'Ticker'])):
    ticker_lookup[key] = group

print("Lookup Created")
```

```python
len(ticker_lookup)
```

```python
LOT_SIZE = 15

def process_trade(row, option_type):

    day = row['Date']
    ticker = row['Ticker']

    option_data = ticker_lookup[(day, ticker)]

    entry_price = row['Close']
    stoploss = entry_price * 1.5

    after_entry = option_data[
        option_data['Time'] > '09:20:59'
    ]

    sl_hit = after_entry[
        after_entry['High'] >= stoploss
    ]

    if len(sl_hit) > 0:

        exit_time = sl_hit.iloc[0]['Time']
        exit_price = stoploss

    else:

        exit_row = option_data[
            option_data['Time'] == '15:20:59'
        ]

        if len(exit_row) == 0:
            return None

        exit_time = '15:20:59'
        exit_price = exit_row.iloc[0]['Close']

    pnl = (entry_price - exit_price) * LOT_SIZE

    return {
        'Date': day,
        'Ticker': ticker,
        'OptionType': option_type,
        'EntryTime': '09:20:59',
        'ExitTime': exit_time,
        'EntryPrice': round(entry_price, 2),
        'ExitPrice': round(exit_price, 2),
        'PnL': round(pnl, 2)
    }
```

```python
process_trade(
    ce_trades.iloc[0],
    'CE'
)
```

```python
all_trades = []

for _, row in tqdm(
    ce_trades.iterrows(),
    total=len(ce_trades)
):
    trade = process_trade(row, 'CE')

    if trade:
        all_trades.append(trade)

for _, row in tqdm(
    pe_trades.iterrows(),
    total=len(pe_trades)
):
    trade = process_trade(row, 'PE')

    if trade:
        all_trades.append(trade)

trade_df = pd.DataFrame(all_trades)

print(trade_df.shape)
```

```python
trade_df = trade_df.sort_values(
    ['Date', 'OptionType']
).reset_index(drop=True)

trade_df.head()
```

```python
trade_df['CumulativePnL'] = (
    trade_df['PnL'].cumsum()
)

trade_df.head()
```

```python
INITIAL_CAPITAL = 100000

trade_df['AvailableCapital'] = (
    INITIAL_CAPITAL +
    trade_df['CumulativePnL']
)

trade_df.head()
```

```python
LOT_SIZE = 15

trade_df['Quantity'] = LOT_SIZE

trade_df['EntryValue'] = (
    trade_df['EntryPrice']
    * LOT_SIZE
)

trade_df['ExitValue'] = (
    trade_df['ExitPrice']
    * LOT_SIZE
)
```

```python
trade_df['PnL_%'] = (
    trade_df['PnL']
    /
    trade_df['EntryValue']
) * 100
```

```python
wins = (trade_df['PnL'] > 0).sum()

losses = (trade_df['PnL'] < 0).sum()

total = len(trade_df)

win_rate = wins / total * 100

loss_rate = losses / total * 100

print("Wins :", wins)
print("Losses :", losses)
print("Win Rate :", round(win_rate,2))
print("Loss Rate :", round(loss_rate,2))
```

```python
trade_df['Month'] = (
    pd.to_datetime(
        trade_df['Date']
    ).dt.to_period('M')
)

monthly_pnl = (
    trade_df
    .groupby('Month')['PnL']
    .sum()
)

monthly_pnl
```

```python
daily_pnl = (
    trade_df
    .groupby('Date')['PnL']
    .sum()
)

nav = 100 + daily_pnl.cumsum()

nav.head()
```

```python
start_nav = nav.iloc[0]

end_nav = nav.iloc[-1]

years = 1

cagr = (
    ((end_nav / start_nav) ** (1/years))
    - 1
) * 100

print("CAGR:", round(cagr,2), "%")
```

```python
running_max = nav.cummax()

drawdown = (
    nav - running_max
) / running_max

max_drawdown = (
    drawdown.min() * 100
)

print(
    "Max Drawdown:",
    round(max_drawdown,2),
    "%"
)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))

nav.plot()

plt.title(
    "Equity Curve"
)

plt.grid(True)

plt.show()
```
